## GrÃ¡fico de Pizza

```python
import matplotlib.pyplot as plt
```

```python
help(plt.pie)
```

```python
dir(plt.pie)
```

```python
valores = [15, 30, 45, 10]
rotulos = ['Frogs', 'Hogs', 'Dogs', 'Logs']

plt.figure(figsize=(10,10))

plt.pie(valores, labels=rotulos)
plt.show()
```

```python
valores = [15, 30, 45, 10]
rotulos = ['Frogs', 'Hogs', 'Dogs', 'Logs']
explode = [0, 0.1, 0, 0]
cores = ['blue', 'orange', 'green', 'red']

plt.figure(figsize=(7,7))

plt.pie(
      valores
    , labels=rotulos
    , explode=explode
    #, colors=cores
    , autopct='%.1f%%'
    , pctdistance=0.8    
    , shadow=True
    #, frame=True
    , startangle=90 # valores em graus
    , rotatelabels=True
)
plt.show()
```

## GrÃ¡fico de Pontos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(6,4))
```

```python
help(plt.scatter)
```

```python
x = [1.3, 2.4, 3.3, 4.0, 4.4, 5.0]
y = [63, 34, 22, 48, 18, 5]
tamanho = np.array([10, 6, 3, 8, 2, 10]) * 20
cores = ['Red', 'Green', 'Yellow', 'Yellow', 'Green', 'Red']

plt.scatter(
      x, y
    , s= tamanho
    , c= cores
    , marker="P"
    , cmap='Set1'
    , alpha=0.6
)
plt.xticks(np.arange(1.5, 5.1, 0.5))
plt.yticks(range(20,61,10))
plt.show()
```

```python
x1 = np.random.rand(500)
y1 = np.random.rand(500)

plt.figure(figsize=(10,10))

plt.scatter(
    x1,
    y1,
    s=150,
    alpha=0.6 # Fica escuro onde os pontos se encontram
)
plt.plot()
```

```python
# usando o vsvmax e vmin - para legenda (Para tamanhos muito diferentes)

x = [1.3, 2.4, 3.3, 4.0, 4.4, 5.0]
y = [63, 34, 22, 48, 18, 5]
#tamanho = np.array([10, 6, 3, 8, 2, 10]) * 20
tamanho = [1000, 60, 3, 800, 2, 10]
cores = ['Red', 'Green', 'Yellow', 'Yellow', 'Green', 'Red']

plt.scatter(
      x, y
    , s= tamanho
    , c= cores
    , marker="P"
    , cmap='Set1'
    , alpha=0.6
)
plt.xticks(np.arange(1.5, 5.1, 0.5))
plt.yticks(range(20,61,10))
plt.show()
```

## GrÃ¡ficos de Barra

```python
help(plt.bar)
```

```python
x = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
y = [10, 8, 6, 4, 2, 1]
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    x,
    y,
    color='mediumslateblue',
    alpha=0.8,
    edgecolor='black'
)

plt.ylabel('Usage')
plt.title('Programming Language Usage')
plt.ylim(0,11)
plt.show()
```
