```python
valores = [15, 30, 45, 10]
rotulos = ['Frogs', 'Hogs', 'Dogs', 'Logs']

plt.figure(figsize=(10,10))

plt.pie(valores, labels=rotulos)
plt.show()
```

```python
valores = [15, 30, 45, 10]
rotulos = ['Frogs', 'Hogs', 'Dogs', 'Logs']
explode = [0, 0.1, 0, 0]
cores = ['blue', 'orange', 'green', 'red']

plt.figure(figsize=(7,7))

plt.pie(
      valores
    , labels=rotulos
    , explode=explode
    #, colors=cores
    , autopct='%.1f%%'
    , pctdistance=0.8    
    , shadow=True
    #, frame=True
    , startangle=90 # valores em graus
    , rotatelabels=True
)
plt.show()
```
