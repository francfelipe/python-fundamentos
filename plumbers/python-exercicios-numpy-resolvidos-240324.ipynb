### 3. Crie um array com tr√™s dimens√µes, onde a primeira dimens√£o s√£o os dias da semana (seg a dom), a segunda dimens√£o s√£o as semanas do m√™s (considere apenas 4 para todos os meses), e a terceira s√£o os meses do ano (Jan a Dez).

Aqui temos quase uma hierarquia: meses, semanas e dias.
A maior hierarquia √© quem vem primeiro.

```python
np.zeros([12,4,7])
# aqui j√° fica com uma cara de calend√°rio
```

```python
# vamos fazer com vazio pois vamos trabalhar com textos e para o array trabalhar com o formato certo

calendario = np.full([12,4,7], '')
calendario
```

a. Marque os finais de semana com a letra W

```python
# cola para ajudar a filtrar
# [meses. semanas, dias]

calendario[:, :, -2:] = 'W'
calendario
```

b. Marque o come√ßo do m√™s com a letra S

```python
# cola para ajudar a filtrar
# [meses. semanas, dias]

calendario[:, 0, 0].shape
```

```python
calendario[:, 0, 0] = 'S'
calendario
```

c. Marque o final do m√™s com a letra E

```python
# cola para ajudar a filtrar
# [meses. semanas, dias]

calendario[:, -1, -1] = 'E'
calendario
```

d. Marque os demais dias com a letra D

```python
calendario == '' # tudo o que √© vazio

mask_empty = calendario == ''
calendario[mask_empty] = 'D'
calendario
```
